## **a) Population and Sample**

**Population:** The entire group you want to study or draw conclusions about.

**Sample:** A smaller part of the population that you actually collect data from.

**‚úÖ Example**
Suppose you want to find the average height of all students in your school.

**Population:** All students in your school (e.g., 1,000 students).

**Sample:** A group of 100 students that you randomly choose to measure.

## **b) Descriptive Statistics and Inferential Statistics**

**Descriptive Statistics:** Methods for summarizing and organizing data using measures such as mean, median, and standard deviation.

**Inferential Statistics:** Techniques for drawing conclusions or making predictions about a population based on sample data.

| Feature  | Descriptive Statistics          | Inferential Statistics        |
| -------- | ------------------------------- | ----------------------------- |
| Purpose  | Describe data                   | Make predictions/inferences   |
| Based on | Entire data (population/sample) | Usually a sample              |
| Examples | Mean, median, charts            | Hypothesis tests, predictions |


## **c) Parameter and Statistic**

**Parameter:** A numerical value that describes a characteristic of a population (e.g., population mean).

**Statistic:** A numerical value that describes a characteristic of a sample (e.g., sample mean).

| Concept       | Based On   | Known or Unknown   | Example                            |
| ------------- | ---------- | ------------------ | ---------------------------------- |
| **Parameter** | Population | Usually unknown    | Average height of **all** students |
| **Statistic** | Sample     | Known (calculated) | Average height of **100** students |


## **d) Qualitative and Quantitative Data**

**Qualitative Data:** Non-numeric information that describes categories or labels (e.g., colors, names).

**Quantitative Data:** Numeric data that represents measurable quantities (e.g., height, weight).

# **Q2. Measures of Central Tendency - Definitions**

**a) Mean**

Definition: The mean is the average of a set of numbers, calculated by summing all values and dividing by the number of values.

Example: For the numbers 4, 6, and 8: Mean = (4 + 6 + 8) / 3 = 18 / 3 = 6

**b) Median**

Definition: The median is the middle value in an ordered dataset. If there is an even number of values, it's the average of the two middle numbers.

Example: For the numbers 3, 7, 9: Median = 7 (the middle value)

**c) Mode**

Definition: The mode is the value that appears most frequently in a dataset.

Example: For the numbers 2, 4, 4, 6, 8: Mode = 4 (it appears twice)

**Q5. Variance and Standard Deviation - Theory**

**Objective:** Understand spread/variability in data.

**Problem:**

‚óè	a) Define Variance and Standard Deviation.

‚óè	b) Explain why Standard Deviation is more interpretable than Variance.

**üîπ Variance**
Variance measures the average squared difference between each data point and the mean.
It tells us how spread out the data values are.

üìå **Variance Formula:**

$$
\sigma^2 = \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2
$$

**Where:**

ùë•ùëñ  = each data point

ùë•  = mean of the data

n = number of data points

**üîπ Standard Deviation**
Standard deviation is the square root of variance.
It shows the average distance of each data point from the mean.

**üìå Formula:**

**Standard¬†Deviation**
$$
\sigma = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2}
$$


**‚úÖ b) Why is Standard Deviation More Interpretable than Variance?**

üîπ **1. Same Units as the Data:**

Variance is in squared units (e.g., cm¬≤, kg¬≤), which can be hard to interpret.

Standard deviation is in the original units (e.g., cm, kg), making it easier to relate to the data.

üîπ **2. Easier to Understand Magnitude:**

If standard deviation = 5 kg, it means on average, the data points vary by 5 kg from the mean.

If variance = 25 kg¬≤, it‚Äôs not clear what ‚Äú25 kg¬≤‚Äù actually means in real life.

üîπ **3. Common in Practice:**

Standard deviation is more commonly used in statistical summaries, reports, and charts.

**üß† Example:**
If the average test score is 70 and the standard deviation is 10, then:

Most students scored between 60 and 80 (mean ¬± 1 SD).

That range is easy to interpret.

But if the variance is 100, we just know it's "some spread" ‚Äî but not in real, understandable units.

### üîπ Five-number Summary

The **five-number summary** is a set of five descriptive statistics that provide insight into the distribution and spread of a dataset:

| Component        | Description |
|------------------|-------------|
| **Minimum**       | The smallest value in the dataset. |
| **Q1 (First Quartile)** | The 25th percentile ‚Äî 25% of data lies below this value. |
| **Median (Q2)**   | The 50th percentile ‚Äî middle value when data is sorted. |
| **Q3 (Third Quartile)** | The 75th percentile ‚Äî 75% of data lies below this value. |
| **Maximum**       | The largest value in the dataset. |



### üîπ Example of a Five-number Summary

For the dataset:  
`22, 25, 29, 31, 35, 40, 45, 48, 50`

- **Minimum** = 22  
- **Q1** = 27  
- **Median** = 35  
- **Q3** = 46.5  
- **Maximum** = 50  

---

### üîπ Boxplot Concept

A **boxplot** (or box-and-whisker plot) is a visual summary of the five-number summary. It shows:

- A box from **Q1 to Q3** (the interquartile range, IQR)
- A line inside the box for the **median**
- "Whiskers" that extend to the **minimum** and **maximum** (within limits)
- **Outliers** are plotted as individual points beyond the whiskers

---

### üîπ How Boxplots Help in Detecting Outliers

Boxplots detect **outliers** using the IQR rule:

- Any data point:
  - **Below** Q1 ‚àí 1.5 √ó IQR  
  - **Above** Q3 + 1.5 √ó IQR  
  is considered an **outlier**.

These are shown as dots or stars on the plot.

---

### ‚úÖ Summary

- The **five-number summary** provides a quick overview of data distribution.
- A **boxplot** uses this summary visually to show spread, symmetry, and outliers.



**Q9. Confidence Interval for the Mean**

**Objective:** Estimate population means using sample data.

**Problem:**

 A sample of 36 students has an average height of 162 cm with a standard deviation of 6 cm.

 Calculate the 95% Confidence Interval for the population mean.

 (Hint: Use Z = 1.96 for 95% confidence)


### üîπ Problem Statement:

A sample of **36 students** has:
- **Sample Mean (xÃÑ)** = 162 cm  
- **Sample Standard Deviation (s)** = 6 cm  
- **Sample Size (n)** = 36  
- **Confidence Level** = 95%  
- **Z-value for 95%** = 1.96


### üîπ Formula for Confidence Interval (CI):

$$
\text{CI} = \bar{x} \pm Z \cdot \frac{s}{\sqrt{n}}
$$

Where:
- \( \bar{x} \) = sample mean  
- \( Z \) = Z-score for 95% confidence  
- \( s \) = sample standard deviation  
- \( n \) = sample size

---

### üîπ Step-by-Step Calculation:

**1. Calculate Standard Error (SE):**

$$
SE = \frac{s}{\sqrt{n}} = \frac{6}{\sqrt{36}} = \frac{6}{6} = 1
$$

**2. Apply the Confidence Interval formula:**

$$
CI = 162 \pm 1.96 \cdot 1 = 162 \pm 1.96
$$

---

### ‚úÖ Final Answer:

- **Lower Limit**: \( 162 - 1.96 = \boxed{160.04} \)  
- **Upper Limit**: \( 162 + 1.96 = \boxed{163.96} \)

---

### üìå Conclusion:

The **95% Confidence Interval** for the population mean height is:

$$
\boxed{(160.04\ \text{cm},\ 163.96\ \text{cm})}
$$


**Q10. Hypothesis Testing - One Sample Z-Test**

**Objective:** Make decisions using statistical testing.

**Problem:**

 The average salary in a city is ‚Çπ30,000. A random sample of 49 employees has an average salary of ‚Çπ31,000 with a standard deviation of ‚Çπ4,900.
 Test the hypothesis at the 5% level of significance to determine if the average salary has increased.

‚óè	a) State the null and alternative hypothesis:

‚óè	b) Calculate the Z-score:

‚óè	c) Conclude the result using critical value (¬±1.96)




### üìù Problem Summary

- **Population Mean (Œº)** = ‚Çπ30,000  
- **Sample Mean (xÃÑ)** = ‚Çπ31,000  
- **Sample Size (n)** = 49  
- **Sample Standard Deviation (s)** = ‚Çπ4,900  
- **Significance Level (Œ±)** = 0.05  
- **Z-critical (two-tailed, 5%)** = ¬±1.96  

---

### üîπ a) Hypotheses

- **Null Hypothesis (H‚ÇÄ)**:  
  \( \mu = 30,000 \)  
  (The average salary has not increased)

- **Alternative Hypothesis (H‚ÇÅ)**:  
  \( \mu > 30,000 \)  
  (The average salary has increased)

‚Üí This is a **one-tailed (right-tailed)** test.

---

### üîπ b) Calculate the Z-score

**Formula:**

$$
Z = \frac{\bar{x} - \mu}{\frac{s}{\sqrt{n}}}
$$

**Substitute values:**

- \( \bar{x} = 31,000 \)
- \( \mu = 30,000 \)
- \( s = 4,900 \)
- \( n = 49 \)

**Step 1: Compute Standard Error (SE)**

$$
SE = \frac{4{,}900}{\sqrt{49}} = \frac{4{,}900}{7} = 700
$$

**Step 2: Compute Z-score**

$$
Z = \frac{31{,}000 - 30{,}000}{700} = \frac{1{,}000}{700} \approx \boxed{1.43}
$$

---

### üîπ c) Decision and Conclusion

- **Z-calculated** = 1.43  
- **Z-critical** (one-tailed, 5% level) = **1.645**

**Decision Rule:**

- If \( Z > 1.645 \), **reject H‚ÇÄ**
- If \( Z < 1.645 \), **fail to reject H‚ÇÄ**

**Since 1.43 < 1.645**, we **fail to reject the null hypothesis**.

---

### ‚úÖ Final Conclusion:

At the 5% level of significance, there is **not enough evidence** to conclude that the average salary has increased.

**One Sample Z-Test (example: test if average price ‚â† 15000)**

```python
# Hypothesis Test
pop_mean = 15000  # replace with population mean if known
sample_mean = df[numeric_column].mean()
sample_std = df[numeric_column].std()
n = df[numeric_column].count()

z = (sample_mean - pop_mean) / (sample_std / np.sqrt(n))
print(f"\nZ-Score: {z}")

if abs(z) > 1.96:
    print("Reject Null Hypothesis: There is a significant difference from ‚Çπ15,000.")
else:
    print("Fail to Reject Null Hypothesis: No significant difference from ‚Çπ15,000.")
```
